## Import Library

```python
import h5py as hp
import pandas as pd
import numpy as np
import time as tm
```

## Load song_json 

```python
df2=pd.read_csv('song.json',sep='\t')
df2.rename(columns={'id': 'song_id'}, inplace=True)
df2.head()
```

## Load Various_Artist

```python
Various_Artist=pd.read_hdf('Various_Artist.h5','a')
Various_Artist.head()
```

## Merge artist by song_id and Remove Various_Artist

```python
size=0
start_time=tm.time()
for ii in range(97,101,1):
    for j in range(97,123,1):
        file_name='1'+chr(ii)+chr(j)+'.h5'
        df=pd.read_hdf(file_name,'a')
        df=df.merge(df2,on='song_id', how='left')
        df.drop(['song_id','played_at','month','day','week','album_id'], axis=1, inplace=True)
        df=df[~df['artist_id'].isin(Various_Artist.artist_id)]
        df.to_hdf(file_name,'a')
        print ('File_name:%s Data_size:%d'%(file_name,len(df)))
        size=size+len(df)
        if(ii==100 and j==108):
            break
end_time=tm.time()-start_time
print ('Data_size:%d Process_time:%d'%(size,end_time))
```
