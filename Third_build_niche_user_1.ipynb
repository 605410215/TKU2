```python
import h5py as hp
import pandas as pd
import numpy as np
import time as tm
```

## Calculate user hear artist's number

```python
size=0
start_time=tm.time()
columns=['msno','artist_id','number']
dataset2=pd.DataFrame(columns=columns)
dataset2=dataset2.groupby(['msno','artist_id']).count()
for ii in range(97,101,1):
    for j in range(97,123,1):
        file_name='1'+chr(ii)+chr(j)+'.h5'
        df=pd.read_hdf(file_name,'a')
        df=df.groupby(['msno','artist_id']).count()
        dataset2=dataset2.add(df,fill_value=0)
        print ('File_name:%s Data_size:%d'%(file_name,len(df)))
        size=size+len(df)
        if(ii==100 and j==108):
            break
end_time=tm.time()-start_time
print ('Data_size:%d Process_time:%d'%(size,end_time))
```
