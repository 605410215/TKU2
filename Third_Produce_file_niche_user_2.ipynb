## Import Library

```python
import pandas as pd
import h5py
import numpy as np
import math
from scipy.stats import entropy
import time
```

## Load user and artist's network

```python
df=pd.read_hdf('user_artist_network.h5','a')
df
```

## Calculate msno hear total number

```python
df['artist_total_number']=df.groupby(['msno'])['number'].transform("sum")
df
```

## Calculate user hear max rate

```python
df['number_rate']=df['number']/df['artist_total_number']
df['max_rate']=df.groupby("msno")["number_rate"].transform("max")
##
df['max_rate']=df.max_rate.round(2)
df
```

## Calculate entropy for each user

```python
df['each_artist_entrop']=df['max_rate'].apply(lambda n:(-n * np.log2(n))-((1-n)*np.log2(1-n)))
df
```

## Convert entropy equal zero

```python
df2=df.fillna(0)
df2
```

```python
df2['each_artist_entrop']=df2.each_artist_entrop.round(2)
df2=df2.groupby(['each_artist_entrop','max_rate']).count().reset_index()
df2=df2.drop(['artist_id','number','artist_total_number','number_rate'],axis=1)
df2.rename(columns={'msno': 'number'}, inplace=True)
```
